```python
from pyspark import SparkConf, SparkContext
```

```python
sconf = SparkConf().setAppName("WordCounting").setMaster("local[*]")
```

```python
sc = SparkContext(conf = sconf)
```

```python
lines = sc.textFile("text.txt")
```

```python
wordCounts = lines.flatMap(lambda x : x.split())\
                  .map(lambda x: (x,1))\
                  .reduceByKey(lambda x,y:x+y)\
                  .sortBy(lambda x: -x[1])\
                  .collect()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
counts = np.array([list(x) for x in wordCounts])
```

```python
xData = np.arange(0,len(wordCounts))
yData = counts[:,1].astype(int)

theo_data = [np.max(yData)/x for x in np.arange(1,len(wordCounts)+1)]
plt.ylabel = counts[:,0]
plt.plot(xData[0:200],yData[0:200])
plt.plot(xData[0:200],theo_data[0:200])
```
